## Question 2: What US city has the highest number of Sales?

```python
#add city column
#apply method
def getcity(address):
    return address.split(',')[1]
def getstate(address):
    return address.split(',')[2].split(' ')[1]
data['City'] = data['Purchase Address'].apply(lambda x:getcity(x) + ' ' + getstate(x))
#data['City'] = data['Purchase Address'].apply(lambda x: x.split(',')[1]) #without using function
data.head()
```

```python
results = data.groupby('City').sum(numeric_only = True)
results
```

```python
#cities = data['City'].unique() #showing data error 
#so i am going to change a bit
cities = [city for city,df in data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation = 'vertical', size=8)
plt.ylabel('Sales is USD ($)')
plt.xlabel('City Names')
plt.show()
```

#### Question 4: What products are often sold together?

```python
data
```

```python
#to get similar order id
df = data[data['Order ID'].duplicated(keep = False)]
df.head(20)

#grouping everythig together
df['Same_Products'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
df = df[['Order ID', 'Same_Products']].drop_duplicates()
df.head()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()
for row in df['Same_Products']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    #count.update(Counter(combinations(row_list, 3)))

for key,value in count.most_common(10):
    print(key, value)
```
